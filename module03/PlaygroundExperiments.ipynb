## Creating a Dataset
creating an example dataset to run experiemnts over for a prompt

```python
from langsmith import Client

example_inputs = [
("How common of a hobby is cooking?", "Roughly 38% of people identifying it as a hobby"),
("What percent of Gen Z can cook?", "Nearly two thirds of Gen Z adults admit they lack basic cooking skills - with around 61 per cent of those surveyed saying they do not know how to cook an omelette."),
("What are any 2 easy pasta recipes to cook", "Creamy garlic pasta and Creamy pesto and kale pasta"),
]

client = Client()
dataset_name = "dataset questions"

dataset = client.create_dataset(
  dataset_name=dataset_name, description="Sample questions about cooking",
)

inputs = [{"question": input_prompt} for input_prompt, _ in example_inputs]
outputs = [{"output": output_answer} for _, output_answer in example_inputs]

client.create_examples(
  inputs=inputs,
  outputs=outputs,
  dataset_id=dataset.id,
)
```
